##### Mean Absolute Error


```python
# Import Libraries
from sklearn.metrics import mean_absolute_error 
#----------------------------------------------------

y_test = [1,5,3,6,7,8,9]
y_pred = [1,6,3,3,2,8,9]

#Calculating Mean Absolute Error
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average') # it can be raw_values
'''
if multioutput='uniform_average' that mean it will calc the mean of all values 
if multioutput='raw_values' that mean it will calc the mean of each row alone

'''
print('Mean Absolute Error Value is : ', MAEValue)
```

#### confusion_matrix

```python
#Import Libraries
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
#----------------------------------------------------

y_test = [1,0,0,1,1,1]
y_pred = [1,0,0,0,1,0]

#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

'''
             pred 1  | pred 0
 -----------------
  actual 1   | TP     | FP
 -----------------
  actual 0   | FN     | TN
  
'''
'''
             pred 1  | pred 0
 -----------------
  actual 1   | 2     | 0
 -----------------
  actual 0   | 2     | 2
  
'''

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```

#### F1 Score
F1 score is a measure of a model's accuracy that takes into account both the precision and recall of the model.

The precision of a model is the proportion of predicted positive labels that are actually positive. The recall of a model is the proportion of actual positive labels that are predicted as positive. The F1 score is calculated as the harmonic mean of precision and recall.

The F1 score can range between 0 and 1, with 0 being the worst score and 1 being the best score. A score of 0.5 indicates that the model is predicting the positive class with equal precision and recall.

```python
#Import Libraries
from sklearn.metrics import f1_score
#----------------------------------------------------

#Calculating F1 Score  : 2 * (precision * recall) / (precision + recall)
# f1_score(y_true, y_pred, labels=None, pos_label=1, average=’binary’, sample_weight=None)

F1Score = f1_score(y_test, y_pred, average='micro') #it can be : binary,macro,weighted,samples
'''
'binary': Only report the F1 score for the positive class. This is only applicable for binary classification problems.
'micro': Calculate the F1 score globally, by counting the total true positives, false positives, and false negatives. This is the most common way to calculate the F1 score for multiclass problems.
'macro': Calculate the F1 score for each class, and then take the unweighted average. This does not take into account class imbalance.
'weighted': Calculate the F1 score for each class, and then take the weighted average, where the weights are the number of true instances for each class.
'''
print('F1 Score is : ', F1Score)
```

#### Recall Score
. The recall score is a measure of a model's ability to find all of the positive instances in a dataset.

The recall score is calculated as the ratio of the number of true positives to the sum of the number of true positives and the number of false negatives. 

A perfect recall score is 1, and a score of 0 indicates that the model did not find any of the positive instances.

```python
#Import Libraries
from sklearn.metrics import recall_score
#----------------------------------------------------


#Calculating Recall Score : (Sensitivity) (TP / float(TP + FN))   1 / 1+2  
# recall_score(y_true, y_pred, labels=None, pos_label=1, average=’binary’, sample_weight=None)

RecallScore = recall_score(y_test, y_pred, average='micro') #it can be : binary,macro,weighted,samples
'''
average: The type of averaging to use. The following values are supported:
'binary': Only report the recall score for the positive class. This is only applicable for binary classification problems.

'micro': Calculate the recall score globally, by counting the total true positives, false positives, and false negatives. T
his is the most common way to calculate the recall score for multiclass problems.

'macro': Calculate the recall score for each class, and then take the unweighted average. This does not take into account class imbalance.

'weighted': Calculate the recall score for each class, and then take the weighted average, 
where the weights are the number of true instances for each class. This alters 'macro' to account for class imbalance.

None: Return the recall score for each class.
'''
print('Recall Score is : ', RecallScore)
```

#### precision_score
calculates the precision of a binary or multiclass classification. Precision is the ratio of true positives to the sum of true positives and false positives.

 In other words, it is the fraction of predicted positive instances that are actually positive.


```python
#Import Libraries
from sklearn.metrics import precision_score
#----------------------------------------------------

#Calculating Precision Score : (Specificity) #(TP / float(TP + FP))  
# precision_score(y_true, y_pred, labels=None, pos_label=1, average=’binary’,sample_weight=None)
y_pred =  ['a','b','c','a','b','c','a','b','c','a']
y_true =  ['a','a','b','b','a','b','c','c','b','b']
PrecisionScore = precision_score(y_true, y_pred, average='micro') 
''' 
average parameter determines how the precision score is calculated for multiclass classification. The following averaging methods are available:

'binary': Only the precision for the positive class is returned. This is only applicable when the target is binary.

'micro': The precision is calculated globally, by summing the true positives
and false positives for each class and dividing by the sum of the true positives for all classes.

'macro': The precision is calculated for each class and then averaged. 
The macro average does not take into account the class imbalance.

'weighted': The precision is calculated for each class and then averaged,
weighted by the number of samples in each class. The weighted average takes into account the class imbalance.

'''
print('Precision Score is : ', PrecisionScore)
```

#### roc_curve and auc
It is a tool for determining the right value for sensitivity & specificity,

 and selecting the right threshold, noting that we give it y_pred_prob it takes potential and not expected values

```python

import numpy as np
from sklearn import metrics
y =      np.array([1    , 1     , 2     , 2])
scores = np.array([0.1  , 0.4   ,   0.35, 0.8])
fpr, tpr, thresholds = metrics.roc_curve(y, scores, pos_label=2)
print(fpr,tpr)
```

* AUC

he AUC tool is the acronym area under curve They calculate the space under the previous curve explained,

 and note that the larger the area under the curve the more accuracy of the algorithm indicates, because it allows high sense & specificity values together.


```python
#Import Libraries
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import numpy as np
#----------------------------------------------------

#Calculating Area Under the Curve :  
y =      np.array([1    , 1     , 2     , 2])
scores = np.array([0.1  , 0.4   ,   0.35, 0.8])

fprValue2, tprValue2, thresholdsValue2 = roc_curve(y,scores,pos_label=2)
AUCValue = auc(fprValue2, tprValue2)
print('AUC Value  : ', AUCValue)
```

```python
import numpy as np
from sklearn import metrics
y =      np.array([1    , 1     , 2     , 2])
scores = np.array([0.1  , 0.4   ,   0.35, 0.8])
fpr, tpr, thresholds = metrics.roc_curve(y, scores, pos_label=2)

metrics.auc(fpr, tpr)
```
